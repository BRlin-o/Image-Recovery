```python
## mask get

import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./data/parrot.jpg")
print(img[0:5, 0:5])
plt.imshow(img)
plt.show()


img_defeat = cv2.imread('./data/parrot_defeat.PNG')
print(img_defeat[0:5, 0:5])
plt.imshow(img_defeat)
plt.show()
```

```python
mask = np.array(img - img_defeat)
plt.imshow(mask)
plt.show()
new_mask = cv2.bilateralFilter(mask, 0, 255, 5)
plt.imshow(new_mask)
plt.show()
## cv2.imwrite('./data/parrot_mask.png', mask)
```
